# 4) Changing datatypes and adding new columns

```python
# Changing datatype of column from object to datestamp

bikes['started_at'] = pd.to_datetime(bikes['started_at'])
```

```python
# Changing datatype of column from object to datestamp

bikes['ended_at'] = pd.to_datetime(bikes['ended_at'])
```

```python
# Checking for changes made in datatypes of each column

bikes.dtypes
```

```python
# Creating column for how much time spent on bike ride, labeled "time_spent_with_bike"
# Formula was derived from taking the difference between "ended_at" and "started_at" column

bikes["time_spent_with_bike"] = bikes['ended_at'] - bikes['started_at']
bikes
```

```python
# Changed data type of 'time_spent_with_bike' column to an integer and also converted time spent to total seconds

bikes['time_spent_with_bike'] = bikes['time_spent_with_bike'].dt.total_seconds().astype(int)
bikes
```

```python
# Created new column derived from 'time_spent_with_bike'
# Converted time spent from total seconds to minutes (nearest two places after decimal)

bikes['time_spent_with_bike_in_minutes'] = round((bikes['time_spent_with_bike']/60),2)
bikes
```

```python
# Created two new columns derived from 'ended_at_day_name', 'started_at_day_name'
# Returned the name of day

bikes['started_at_day_name'] = bikes['started_at'].dt.day_name()
bikes['ended_at_day_name'] = bikes['ended_at'].dt.day_name()
bikes
```

```python
# Created two new columns derived from 'ended_at_day_name', 'started_at_day_name'
# Returned the hour of day

bikes['started_at_hour'] = bikes['started_at'].dt.hour
bikes['ended_at_hour'] = bikes['ended_at'].dt.hour
bikes
```

```python
# Created two new columns derived from 'ended_at_day_name', 'started_at_day_name'
# Returned the month of year

bikes['started_at_month_name'] = bikes['started_at'].dt.month_name()
bikes['ended_at_month_name'] = bikes['ended_at'].dt.month_name()

bikes
```

```python
# Created two new columns derived from 'ended_at_day_name', 'started_at_day_name'
# Returned the year

bikes['started_at_year'] = bikes['started_at'].dt.year
bikes['ended_at_year'] = bikes['ended_at'].dt.year
bikes
```

```python
# Created two new columns derived from 'ended_at_day_name', 'started_at_day_name'
# Returned the year-month-day

bikes['started_at_date'] = bikes['started_at'].dt.to_period('D')
bikes['ended_at_date'] = bikes['ended_at'].dt.to_period('D')
bikes
```

# 5) Outliers

```python
# Statistical description of dataframe
# Unusually the bike ride lasts over 50000 minutes
# There is a monetary penalty for rides longer than 45 minutes for members and 30 minutes for casual customers, it is improbable than any rides will much longer. The System host site indicates that for distance calculation, trips are capped at a 2 hour limit and we will use a similar approach.
# It is highly probable that the bike was not engaged properly at the returning station.

bikes.describe()
```

```python
# We will filter out any trips greater than 2 hours.
bikes = bikes[bikes['time_spent_with_bike_in_minutes']<=120]
bikes
```
