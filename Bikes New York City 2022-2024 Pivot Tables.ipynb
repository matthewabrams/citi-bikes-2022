# 1) Importing Libraries and DataFrame

```python
# Importing libraries

import pandas as pd
import numpy as np
import os
from datetime import datetime
```

```python
# Importing bikes_2023-2024.csv

path = r'/Users/matthewabrams'
nyc_bikes_2023= pd.read_csv(os.path.join(path, 'Desktop', 'bikes_2023-2024.csv'), index_col = False)
nyc_bikes_2023.drop(columns = ['Unnamed: 0'], inplace = True)
nyc_bikes_2023
```

```python
# Importing bikes_2022-2023.csv

nyc_bikes_2022= pd.read_csv(os.path.join(path, 'Desktop', 'bikes_2022-2023.csv'), index_col = False)
nyc_bikes_2022.drop(columns = ['Unnamed: 0'], inplace = True)
nyc_bikes_2022
```

```python
# Combining all dataframes for nyc using concat method (historical data in 2023)

frames = [nyc_bikes_2023, nyc_bikes_2022]

nyc_bikes= pd.concat(frames, axis = 0)
nyc_bikes 
```

# 2) Aggregating Data using Pivot Tables

```python
# Aggregating number of bike rides by count grouped by time and location
time = pd.pivot_table(data = nyc_bikes, values = 'time_spent_with_bike_in_minutes', index = ['started_at_year','started_at_month_name', 'started_at_day_name',  
       'started_at_hour', 'end_station_name', 'member_casual', 'end_lat', 'end_lng'], aggfunc ='count', margins = True, margins_name = 'All').astype(int)
time
```

```python
# Aggregating number of locations where bike rides are biked to by count
location = pd.pivot_table(data = nyc_bikes, values = 'time_spent_with_bike_in_minutes', 
                          index = ('end_station_name', 'end_lat', 'end_lng'), 
                          aggfunc ='count', margins = True, margins_name = 'All')
location.fillna(0,inplace = True)
location.astype(int)
```

# 3) Exporting pivot tables as csvs

```python
# Exporting pivot tables as csvs

time.to_csv(os.path.join(path, 'Desktop', 'bike_time.csv'))
location.to_csv(os.path.join(path, 'Desktop', 'bike_location.csv'))
```
