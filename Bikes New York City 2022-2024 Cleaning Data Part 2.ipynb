# 6) Changing datatypes and adding new columns

```python
# Changing datatype of column from object to datestamp

nyc_bikes_2023['started_at'] = pd.to_datetime(nyc_bikes_2023['started_at'])
```

```python
nyc_bikes_2022['started_at'] = pd.to_datetime(nyc_bikes_2022['started_at'])
```

```python
# Changing datatype of column from object to datestamp

nyc_bikes_2023['ended_at'] = pd.to_datetime(nyc_bikes_2023['ended_at'])
```

```python
nyc_bikes_2022['ended_at'] = pd.to_datetime(nyc_bikes_2022['ended_at'])
```

```python
# Checking for changes made in datatypes of each column

nyc_bikes_2023.dtypes
```

```python
nyc_bikes_2022.dtypes
```

```python
# Creating column for how much time spent on bike ride, labeled "time_spent_with_bike"
# Formula was derived from taking the difference between "ended_at" and "started_at" column

nyc_bikes_2023["time_spent_with_bike"] = abs(nyc_bikes_2023['ended_at'] - nyc_bikes_2023['started_at'])
nyc_bikes_2023 
```

```python
nyc_bikes_2022["time_spent_with_bike"] = abs(nyc_bikes_2022['ended_at'] - nyc_bikes_2022['started_at'])
nyc_bikes_2022 
```

```python
# Changed data type of 'time_spent_with_bike' column to an integer and also converted time spent to total seconds

nyc_bikes_2023['time_spent_with_bike'] = nyc_bikes_2023['time_spent_with_bike'].dt.total_seconds().astype(int)
nyc_bikes_2023 
```

```python
nyc_bikes_2022['time_spent_with_bike'] = nyc_bikes_2022['time_spent_with_bike'].dt.total_seconds().astype(int)
nyc_bikes_2022
```

```python
# Created new column derived from 'time_spent_with_bike'
# Converted time spent from total seconds to minutes (nearest two places after decimal)

nyc_bikes_2023['time_spent_with_bike_in_minutes'] = round((nyc_bikes_2023['time_spent_with_bike']/60),2)
nyc_bikes_2023 
```

```python
nyc_bikes_2022['time_spent_with_bike_in_minutes'] = round((nyc_bikes_2022['time_spent_with_bike']/60),2)
nyc_bikes_2022 
```

```python
nyc_bikes_2022['started_at_day_name'] = nyc_bikes_2022['started_at'].dt.day_name()
nyc_bikes_2022['ended_at_day_name'] = nyc_bikes_2022['ended_at'].dt.day_name()
nyc_bikes_2022
```

```python
# Created two new columns derived from 'ended_at_day_name', 'started_at_day_name'
# Returned the hour of day

nyc_bikes_2023['started_at_hour'] = nyc_bikes_2023['started_at'].dt.hour
nyc_bikes_2023['ended_at_hour'] = nyc_bikes_2023['ended_at'].dt.hour
nyc_bikes_2023
```

# 7) Filtering out outliers and rides longer than 2 hrs (120 minutes)

```python
# Statistical description of dataframe
# Unusually the bike ride lasts over 50000 minutes
# There is a monetary penalty for rides longer than 45 minutes for members and 30 minutes for casual customers
# It is less likely that any citi bike ride would take several hours
# The System host site indicates that for distance calculation, trips are capped at a 2 hour (120 minutes) limit
# It is highly probable that the bike was not engaged properly at the returning station

nyc_bikes_2023.describe()
```

```python
nyc_bikes_2022.describe()
```

```python
# We will filter out any trips greater than 2 hours (120 minutes) and any values expressed as negative numbers
nyc_bikes_2023 = nyc_bikes_2023[(nyc_bikes_2023['time_spent_with_bike_in_minutes']<=120) & (nyc_bikes_2023['time_spent_with_bike_in_minutes']>=0)]
nyc_bikes_2023
```

```python
nyc_bikes_2022 = nyc_bikes_2022[(nyc_bikes_2022['time_spent_with_bike_in_minutes']<=120) & (nyc_bikes_2022['time_spent_with_bike_in_minutes']>=0)]
nyc_bikes_2022
```
